```python
data1["brand"]=data1.car_name.apply(lambda x: x.split()[0])
```

```python
data1["price"]=data1.price.apply(lambda x:x.split(".")[0] if "." in x else x)
data1["price"]=data1.price.apply(lambda x: int(x.replace("K",""))*1000 if "K" in x else int(x))
data1.price.value_counts()
```

```python
data1["brand"]=data1.brand.apply(lambda x: "volkswagen" if x in ["vw",'vokswagen'] else x)
data1["brand"]=data1.brand.apply(lambda x:"Mazda" if x in ["maxda","mazda"] else x)
data1["brand"]=data1.brand.apply(lambda x:"Nissan" if x=="nissan" else x)
data1["brand"]=data1.brand.apply(lambda x:"Porsche" if x in ["porcshce",'porsche'] else x)
data1["brand"]=data1.brand.apply(lambda x: "Toyota" if x in ["toyota",'toyouta'] else x)
```

```python
data1.brand.unique()
```

```python
data1["car_name"]=data1.car_name.apply(lambda x: x.replace("vw","volkswagen") if "vw" in x else x)
data1["car_name"]=data1.car_name.apply(lambda x: x.replace('vokswagen',"volkswagen") if 'vokswagen' in x else x)
data1["car_name"]=data1.car_name.apply(lambda x: x.replace("maxda","Mazda") if "maxda" in x else x)
data1["car_name"]=data1.car_name.apply(lambda x: x.replace("mazda","Mazda") if "mazda" in x else x)
data1["car_name"]=data1.car_name.apply(lambda x: x.replace("nissan","Nissan") if "nissan" in x else x)
data1["car_name"]=data1.car_name.apply(lambda x: x.replace("porcshce","Porsche") if "porcshce" in x else x)
data1["car_name"]=data1.car_name.apply(lambda x: x.replace("porsche","Porsche") if "porsche" in x else x)
data1["car_name"]=data1.car_name.apply(lambda x: x.replace("toyota","Toyota") if "toyota" in x else x)
data1["car_name"]=data1.car_name.apply(lambda x: x.replace('toyouta',"Toyota") if 'toyouta' in x else x)
data1.head()
```
